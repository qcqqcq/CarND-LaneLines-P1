Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img,lines

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

````python
# Created this function to filter just yellow and white lines but ended up not using it
def get_yellow_white_mask(image):
    '''Select only white and yellow from images.'''
    
    # Conver to HSV
    hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    #plt.imshow(hsv)
    
    # Define range of yellow and white color in HSV
    lower_white = np.array([0,0,200])
    upper_white = np.array([180,25,255])

    lower_yellow = np.array([20,50,100])
    upper_yellow = np.array([30,255,255])

    # Threshold the HSV image
    mask_yellow = cv2.inRange(hsv, lower_yellow, upper_yellow)    
    mask_white = cv2.inRange(hsv, lower_white, upper_white)

    # Combine yellow OR white
    mask = cv2.bitwise_or(mask_yellow, mask_white)

    #plt.imshow(mask)
    return mask
````

````python
def detect_lines(image):
    '''Detect lines using Canny edge detection and Hough transform'''

    
    # Convert to gray
    gray = grayscale(image)
    #plt.imshow(gray,cmap='gray')
    
    # Apply Gaussian blur
    blur = gaussian_blur(gray, 7)
    #plt.imshow(blur,cmap='gray')
    
    # Do Canny edge detection    
    canned = canny(blur, 50, 150)
    #plt.figure()
    #plt.imshow(canned)
    
    # Keep only lines within the ROI
    # To experiment with other vertices for ROI, use seperate sandbox area
    vertices = np.array([[(100,540),(425, 325), (550, 325), (950,540)]], dtype=np.int32)
    roi = region_of_interest(canned,vertices)

    # Apply Hough transform to detect lines
    rho, theta, threshold, min_line_len, max_line_gap = 2,np.pi/180,15,10,10
    houghed,lines = hough_lines(roi, rho, theta, threshold, min_line_len, max_line_gap)
    #plt.imshow(houghed)
    
    # Return both overlayed image and raw detected lines
    return houghed,lines
````

````python
def get_average_lines(lines):
    '''Given raw lines, apply rules to estimate 1 left and 1 right lane'''
    
    # Calculate characteristics of each line
    # Line shape is (# lines) x 1 x 4 where the 4 values are [x1 y1 x2 y2]
    delta_y = lines[:,0,3] - lines[:,0,1] # y2 - y1
    delta_x = lines[:,0,2] - lines[:,0,0] # x2 - x1
    slopes = (delta_y / delta_x).astype(float) # Slope is dy/dx
    lengths = (delta_y**2 + delta_x**2)**0.5 # Pythagreas
    intercepts = lines[:,0,3] - slopes*lines[:,0,2] # From y = mx + b, solve b = y - mx

    # Choose cut-off values for low and high slope
    low_slope = 0.3 # any slopes less than this are "too horizontal" to be lane markers
    high_slope = 5  # any slopes geater than this are "too vertical" to be lane markers
    valid = (abs(slopes) > low_slope) & (abs(slopes) < high_slope) # keep only reasonably-sloped lines

    # Determine left vs right (using sign of slope)
    right_lines = valid & (slopes > 0)
    left_lines = valid & (slopes < 0)
    
    # Lanes will be drawn fom bottom of screen to some "highest_y" height.
    # Choose highest_y to be highest detected line (where "highest" is lowest y value)
    # This is for costmetic reasons so lines are not drawn until intersecting at vanishing point
    highest_y = min(min(lines[(left_lines) | (right_lines)][:,0,1])
                    ,min(lines[(left_lines) | (right_lines)][:,0,3]))
    highest_y += 25 # lower by some pixels to prevent any chance of displayed lines intersecting

    # Calculate weighted average line so there's only 1 line for left and 1 line for right lane
    avg_lines = [] # Fill with left line, then right line
    for rl_lines in [left_lines,right_lines]:
        
        # Avg line is weighted by line length (longer is better) 
        # and height (bottom of screen is better).  y1 is used to determine height
        weights = 10*lengths[rl_lines] + (lines[:,0,1]-highest_y)[rl_lines]
        avg_slope = np.average(slopes[rl_lines],weights=weights)
        avg_int = np.average(intercepts[rl_lines],weights=weights)

        y1 = image.shape[0] # Display line to bottom of screen
        y2 = highest_y # Display line from somewhere below intersection of lines
        x1 =(y1 - avg_int)/avg_slope # From y = mx + b, solve: x = (y-b)/m
        x2 =(y2 - avg_int)/avg_slope

        avg_lines.append([x1,y1,x2,y2])
    
    # Format and return
    avg_lines = np.array(avg_lines).astype(int).reshape(2,1,4)
    return avg_lines

# Remainin code in this cell is for testing and QA of the function
img = image_list[0]
image = mpimg.imread(os.path.join('test_images',img))
houghed,lines = detect_lines(image)

avg_lines = get_average_lines(lines)
draw_lines(image, avg_lines , color=[255, 0, 0], thickness=10)
#draw_lines(image, lines[left_lines], color=[255, 0, 0], thickness=2)
plt.imshow(image)
````
